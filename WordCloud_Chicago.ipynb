```
import nltk
import nltk.corpus
```

```
from urllib.request import urlopen
speech = urlopen('https://www.business-standard.com/article/current-affairs/full-text-of-swami-vivekananda-s-chicago-speech-of-1893-117091101404_1.html')
html = speech.read()
print(len(html))
tokens = [tok for tok in html.split()]
```

```
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity ='all'
```

```
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```
#Create function for plotting wordclouds
def plot_wordcloud(wordcloud):
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```
print (tokens)
```

```
#We clean it a bit
from bs4 import BeautifulSoup
print("Calling BS4")
soup = BeautifulSoup(html, 'html.parser')

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract() 
```

```
tokens = [tok for tok in soup.get_text().split()]
type(tokens)
```

```
#Converting tokens from list to string
tokens_str = str(tokens)
type (tokens_str)
```

```
print(tokens)
```

```
wordcloud = WordCloud().generate(tokens_str)
plot_wordcloud(wordcloud)
```

```
from nltk.probability import FreqDist
fdist = FreqDist(tokens)
print (fdist)
```

```
import operator
%matplotlib inline
Freq_dist_nltk=nltk.FreqDist(tokens)
print(Freq_dist_nltk)
sorted_d = sorted(Freq_dist_nltk.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_d[:25])
Freq_dist_nltk.plot(25, cumulative=False)
```

```
#Creating WordCloud after Removing Stopwords
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
stopwords.add("every")
stopwords.add("will")

wordcloud = WordCloud(stopwords=stopwords, relative_scaling=1.0).generate(soup.get_text())
plot_wordcloud(wordcloud)
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords.words('english')
```

```
# Load stop words
stop_words = stopwords.words('english')

# Show stop words
stop_words[:10]

#Clean the tokens

clean_tokens = tokens[:]
 
stpwrd = stopwords.words('english')
 
for token in tokens:
 
    if token in stopwords.words('english'):
 
        clean_tokens.remove(token)
```

```
print(clean_tokens)
```

```
import operator
%matplotlib inline
Freq_dist_nltk=nltk.FreqDist(clean_tokens)
print(Freq_dist_nltk)
sorted_d = sorted(Freq_dist_nltk.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_d[:25])
Freq_dist_nltk.plot(25, cumulative=False)
```

```
#Conver clean tokens to string
clean_str = str(clean_tokens)
type (clean_str)
```

```
#Create wordcloud of clean string
wordcloud = WordCloud().generate(clean_str)
plot_wordcloud(wordcloud)
```

```
#Downloading and importing POS tagging
nltk.download('averaged_perceptron_tagger')

from nltk import pos_tag
Pos_tag = pos_tag(clean_tokens)
print(Pos_tag)
```

```
#Importing porter stemmer
from nltk.stem import PorterStemmer
 
stemmer = PorterStemmer()
Pstem = stemmer.stem(clean_str)

print(Pstem)
```

```
#Create wordcloud of porter stemmer
wordcloud = WordCloud().generate(Pstem)
plot_wordcloud(wordcloud)
```

```
#Importing Snowball stemmer
from nltk.stem import SnowballStemmer
 
stemmer = SnowballStemmer('english')
SBstem = stemmer.stem(clean_str)

print(SBstem)
```

```
# Downloading and Importing Word Net Lemmatizer
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```
#Create Lemmatizer for  Words
lemma = lemmatizer.lemmatize(clean_str, pos="v")
print(lemma)
```

```
#Create wordcloud of  Lemmatized tokens
wordcloud = WordCloud().generate(lemma)
plot_wordcloud(wordcloud)
```

```
#Import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# create the transform
vectorizer = CountVectorizer()
```

```
lemma = [lemma]
```

```
# tokenize and build vocab
vectorizer.fit(lemma)
# summarize
print(vectorizer.vocabulary_)
# encode document
vector = vectorizer.transform(lemma)
# summarize encoded vector
print(vector.shape)
print(type(vector))
print(vector.toarray())
```
