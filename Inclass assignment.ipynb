```python
#Impoting the necessary library

import pandas as pd
import numpy as np
import sklearn 
import pandas_profiling as pd1
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.cross_validation import train_test_split
```

```python
#To write more than 1 code at once
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
#To read the csv file
wine_data=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=';')
```

```python
wine_data.head(5)
```

```python
wine_data.tail(5)
```

```python
#EDA
eda_report1=pd1.ProfileReport(wine_data)
eda_report1.to_file("wine_eda.html")
eda_report1
```

```python
#DATA Transforamtion - Scaling the data into 0 and 1 range
scld=MinMaxScaler(feature_range=(0,1))
arr_scld=scld.fit_transform(wine_data)
df_scld=pd.DataFrame(arr_scld, columns=wine_data.columns)
```

```python
df_scld.head()
df_scld.describe()
```

```python
#Defining feature
#Dropping the zero corelated variables
x=wine_data.drop(['citric acid','free sulfur dioxide','sulphates','quality'],1)
x.head(5)
```

```python
#Defining label
y=wine_data['quality']
y.head(5)
```

```python
y.tail(5)
```

```python
#split the data into TRain and test dateset
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.75,random_state=1)
```

```python
train_x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Model instantitation
lm=LinearRegression()
```

```python
lm
```

```python
#Fit the model
lm.fit(train_x,train_y)
```

```python
#Predicting the model output
predict_x=lm.predict(test_x)
```

```python
predict_x
```

```python
#Mean Squared Error
np.round(metrics.mean_squared_error(test_y,predict_x),5)
```

```python
print("Mean Sqaured Error MSE is ")
np.round(metrics.mean_squared_error(test_y,predict_x),5)
```

```python
#Mean Absolute Error
np.round(mean_absolute_error(test_y,predict_x),5)
```

```python
print("Mean Absolute Error MAE is ")
np.round(mean_absolute_error(test_y,predict_x),5)
```

```python
test_x.head()
```

```python
#Merging predicted quality into the original data and computing the error in estimation for the test data.
fdf=pd.concat([test_x,test_y],1)
fdf['Predicted']=np.round(predict_x,1)
fdf['Prediction_Error']=fdf['quality']-fdf['Predicted']
```
